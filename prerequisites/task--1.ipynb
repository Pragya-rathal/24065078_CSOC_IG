```python
mean = np.mean(x_train, axis=0)
std = np.std(x_train, axis=0)

x_train = (x_train - mean) / std
x_test = (x_test - mean) / std
```

```python

class scratchlr:
    
    def __init__(self,learning_rate,epochs):
        
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs

    def dot(self,x_train):
        predictions = []
        for i in range(x_train.shape[0]):
            total = 0
            for k in range (x_train.shape[1]):
                total = total + x_train.iat[i,k]* self.coef_[k] 
            predictions.append(total)
        return predictions
   
    def mean(self,subtract):
        mean = 0
        for j in subtract:
            mean  = mean + j
        return mean/len(subtract)
        
    def fit(self,x_train,y_train):
        self.intercept_ = 0
        self.coef_ = [1.0 for _ in range(x_train.shape[1])]
        
        for l in range(self.epochs):
            y_hat_initial = self.dot(x_train)
            y_hat = [val + self.intercept_ for val in y_hat_initial]
            
            errors = [y_train.iloc[o] - y_hat[o] for o in range(len(y_train))]
            intercept_der = -2 * self.mean(errors)
            self.intercept_ = self.intercept_ - (self.lr * intercept_der) 
            for s in range(len(self.coef_)):
                coef_der = 0
                for v in range(x_train.shape[0]):
                    coef_der += -2 * errors[v] * x_train.iat[v, s]
                coef_der /= x_train.shape[0]
                self.coef_[s] = self.coef_[s] - (self.lr * coef_der)
        
        print(self.intercept_,self.coef_)
    
    def predict(self,x_test):
        y_pred = self.dot(x_test)
        return [y + self.intercept_ for y in y_pred]
```

```python
lr = scratchlr(learning_rate = .20,epochs =250)
```

```python

lr.fit(x_train,y_train)

y_pred = lr.predict(x_test)

```
