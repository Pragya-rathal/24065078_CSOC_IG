```python
x = df.iloc[:,2:12]
y = df.iloc[:,13]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = .2, random_state = 53)
```

```python
import time
```

```python
import numpy as np
```

```python
mean = np.mean(x_train, axis=0)
std = np.std(x_train, axis=0)

x_train = (x_train - mean) / std
x_test = (x_test - mean) / std
```

```python

class scratchlr:
    
    def __init__(self,learning_rate=0.01,epochs=100):
        
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        
    def fit(self,x_train,y_train):
        self.intercept_ = 0
        self.coef_ = np.ones(x_train.shape[1])
        for i in range(self.epochs):
            y_hat = np.dot(x_train,self.coef_) + self.intercept_
            intercept_der = -2 * np.mean(y_train - y_hat)
            self.intercept_ = self.intercept_ - (self.lr * intercept_der)
            coef_der = -2 * np.dot((y_train - y_hat),x_train)/x_train.shape[0]
            self.coef_ = self.coef_ - (self.lr * coef_der)
        
        print(self.intercept_,self.coef_)
    
    def predict(self,x_test):
        return np.dot(x_test,self.coef_) + self.intercept_
```

```python
lr = scratchlr(learning_rate = .20,epochs =250)
```

```python
start_time = time.time()
lr.fit(x_train,y_train)

y_pred = lr.predict(x_test)
end_time = time.time()
print(end_time-start_time)
```

```python
def mean_absolute_score(y_test, y_pred):
    error = (y_test - y_pred).abs()
    return error.mean()

mean_absolute_score(y_test, y_pred)
```

```python
def mean_squared_error(y_test, y_pred):
   error2 = (y_test - y_pred)**2
   return error2.mean()
mean_squared_error(y_test, y_pred)
```
