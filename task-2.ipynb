```python
import time
```

```python
import numpy as np
```

```python
start_time = time.time()
class scratchLR:
    
    def __init__(self):
        self.coef_ = None
        self.intercept_ = None

    def fit(self, x_train, y_train):
        X_train = np.insert(x_train, 0, 1, axis=1)
        coefs = np.linalg.inv(np.dot(X_train.T, X_train)).dot(X_train.T).dot(y_train)
        self.intercept_ = coefs[0]
        self.coef_ = coefs[1:]

    def predict(self, x_test):
        y_pred = np.dot(x_test, self.coef_) + self.intercept_
        return y_pred

lr = scratchLR()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
end_time = time.time()
t1 = end_time - start_time
print(t1)
```

```python
def mean_absolute_score(y_test, y_pred):
    error = (y_test - y_pred).abs()
    return error.mean()

mean_absolute_score(y_test, y_pred)
```

```python
def mean_squared_error(y_test, y_pred):
   error2 = (y_test - y_pred)**2
   return error2.mean()
mean_squared_error(y_test, y_pred)
```

```python

def r2_score(y_test, y_pred):
    ss_r = ((y_test - y_pred) ** 2).sum()
    ss_m = ((y_test - y_train.mean()) ** 2).sum()
    return (1 - (ss_r / ss_m))
r2_score(y_test, y_pred)
```
